## Data Processing

### Separation: X (features) from Y (predict variable)

```python
X = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

### Spliting the Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=0)
```

### Feature Scaling

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

___
## Model Desing

### Creating and training models / Predictions / Evaluating Model

```python
def get_classifier(kernel, c=1, gamma = 'auto'):
    classifier= SVC(kernel=kernel, random_state=0, C= c, gamma = gamma)
    classifier.fit(X_train, y_train)
    return classifier
```

```python
def get_confusion_matrix(y_test,predictions):
    cm = confusion_matrix(y_true=y_test, y_pred= predictions)
    df_cm =pd.DataFrame(data=cm,index=['Real False', 'Real True'], columns=['Pred False', 'Pred True'])
    return df_cm
```

```python
def plot_classifier(X_set, y_set, title, xlabel, ylabel, classifier):
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                         np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
    plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha = 0.55, cmap = ListedColormap(('red', 'green')))
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    c = ListedColormap(('red', 'green'))(i), label = j)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```

```python
def classify(kernel, c=1, gamma='auto'):
    classifier= get_classifier(kernel=kernel, c = c, gamma = gamma)
    predictions = classifier.predict(X_test)
    print(get_confusion_matrix(y_test, predictions))
    title = 'SVM - ' + kernel.upper() + ' Kernel (Training set)'
    xlabel = 'Age'
    ylabel = 'Purshesed'
    plot_classifier(X_train, y_train, title, xlabel, ylabel, classifier)
    # ***************************
    title = 'SVM - '+ kernel.upper() +' Kernel (Test set)'
    xlabel = 'Age'
    ylabel = 'Purshesed'
    plot_classifier(X_test, y_test, title, xlabel, ylabel, classifier)
```

###  Kernel RBF

```python
classify(kernel='rbf')
```

### Kernel Poly

```python
classify(kernel='poly')
```

### Kernel Sigmoid

```python
classify(kernel='sigmoid')
```
